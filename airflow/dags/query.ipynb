```python
import sqlite3

def get_table_columns(db_path, table_name):
    conn = sqlite3.connect(db_path)
    cursor = conn.cursor()

    try:
        cursor.execute(f"PRAGMA table_info({table_name});")
        columns = cursor.fetchall()
        print(f"Cấu trúc bảng {table_name}:")
        for col in columns:
            # col[1] là tên cột, col[2] là kiểu dữ liệu
            print(f" - {col[1]} ({col[2]})")
    except sqlite3.Error as e:
        print(f"Lỗi khi lấy cấu trúc bảng: {e}")
    finally:
        cursor.close()
        conn.close()

if __name__ == "__main__":
    db_path = "crypto_data.db"
    table_name = "crypto_raw"  # Thay bằng tên bảng cần xem
    get_table_columns(db_path, table_name)
```

```python
import sqlite3

def query_crypto_data(db_path, query):
    # Kết nối tới SQLite database
    conn = sqlite3.connect(db_path)
    cursor = conn.cursor()

    try:
        cursor.execute(query)
        # Lấy tất cả kết quả
        rows = cursor.fetchall()

        # In ra kết quả
        for row in rows:
            print(row)

    except sqlite3.Error as e:
        print(f"Lỗi truy vấn: {e}")

    finally:
        cursor.close()
        conn.close()

if __name__ == "__main__":
    db_path = "crypto_data.db"
    sql_query = "SELECT * FROM crypto_raw ORDER BY Date DESC LIMIT 10;"  # Thay bằng truy vấn SQL bạn muốn thực hiện

    query_crypto_data(db_path, sql_query)
```
